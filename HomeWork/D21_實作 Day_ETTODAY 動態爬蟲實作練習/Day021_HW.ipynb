## 作業目標

根據範例：

1. 取出今天所有的新聞
2. 取出現在時間兩小時內的新聞
3. 根據範例，取出三天前下午三點到五點的新聞

```python
# 打開瀏覽器
from selenium import webdriver
from bs4 import BeautifulSoup

browser = webdriver.Chrome(executable_path='../chromedriver')

browser.get("https://www.ettoday.net/news/news-list.htm")
```

```python
# 每個兩秒鐘自動往下滑
import time
for i in range(200):
    time.sleep(2)
    browser.execute_script("window.scrollTo(0, 200000);")
```

```python
# 取得資料，丟到 BeautifulSoup 解析

html_source = browser.page_source
soup = BeautifulSoup(html_source, "html")

for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

```python
# 關閉瀏覽器
browser.quit();
```

## 1. 取出今天所有的新聞

```python
import time
for d in soup.find(class_="part_list_2").find_all('h3'):
    daytime = time.strptime(d.find(class_="date").text, '%Y/%m/%d %H:%M')
    now = time.localtime()
    if(daytime.tm_mon==now.tm_mon and daytime.tm_mday==now.tm_mday and daytime.tm_year==now.tm_year):
        print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

## 2. 取出現在時間兩小時內的新聞

```python
import time
import datetime
now = time.localtime()
print(time.strftime('%Y-%m-%d %H:%M:%SZ', now))
for d in soup.find(class_="part_list_2").find_all('h3'):
    daytime = time.strptime(d.find(class_="date").text, '%Y/%m/%d %H:%M')
    if(daytime.tm_mon==now.tm_mon and daytime.tm_mday==now.tm_mday and daytime.tm_year==now.tm_year):
        start = datetime.datetime.strptime(d.find(class_="date").text, '%Y/%m/%d %H:%M')
        end = datetime.datetime.now()
        time_diff=end-start
        if(((time_diff.total_seconds())/3600)<=2):
            print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

## 3. 根據範例，取出三天前下午三點到五點的新聞

```python
# 打開瀏覽器
from selenium import webdriver
from bs4 import BeautifulSoup
from datetime import timedelta
import datetime

browser = webdriver.Chrome(executable_path='../chromedriver')
today = datetime.date.today()
twoday = datetime.timedelta(days=3)
two_days_ago = today - twoday
browser.get("https://www.ettoday.net/news/news-list-"+str(two_days_ago)+"-0.htm")
# 每個兩秒鐘自動往下滑
import time
for i in range(100):
    time.sleep(2)
    browser.execute_script("window.scrollTo(0, 100000);")
    # 取得資料，丟到 BeautifulSoup 解析

html_source = browser.page_source
soup = BeautifulSoup(html_source, "html")

for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find(class_="date").text, d.find_all('a')[-1].text)
# 關閉瀏覽器
browser.quit();
```

```python
import time
for d in soup.find(class_="part_list_2").find_all('h3'):
    daytime = time.strptime(d.find(class_="date").text, '%Y/%m/%d %H:%M')
    the_day = two_days_ago
    if(daytime.tm_mon==the_day.month and daytime.tm_mday==the_day.day and daytime.tm_year==the_day.year):
        print(d.find(class_="date").text, d.find_all('a')[-1].text)
```
