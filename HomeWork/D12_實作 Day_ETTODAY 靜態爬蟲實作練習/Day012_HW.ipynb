### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html5lib")

for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find(class_="date").text,d.find(class_="tag").text, d.find_all('a')[-1].text)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
news_dict = dict()
data = soup.find(class_="part_list_2").find_all('h3')
for d in data:
    tag = d.find(class_="tag").text
    title =  d.find_all('a')[-1].text
    if tag in news_dict:
        news_dict[tag].append(title)
    else:
        # create a new array in this slot
        news_dict[tag] = [title]
print(news_dict)
```

### ③ 哪一個類別的文章最多

```python
title_len = [0, '']

for n in news_dict:
    if len(news_dict[n]) > (title_len[0]):
        title_len[0] = len(news_dict[n])
        title_len[1] = n
print(title_len)
```
