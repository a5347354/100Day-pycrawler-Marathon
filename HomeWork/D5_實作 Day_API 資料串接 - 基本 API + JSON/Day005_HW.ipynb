## 作業目標

* 請利用 API: https://www.dcard.tw/_api/forums/pet/posts?popular=true 回答下列問題：

1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？
2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」
3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」

```python
import requests
r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
response = r.text

response
```

```python
# 1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？
import json
data = json.loads(response)
count=0 
for d in data:
    count+=1
print('總共回傳'+str(count)+'筆')
# 每一筆資料包含哪些欄位？
data[0].keys()
```

```python
# 2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」
for d in data:
    print('標題'+d['title'])
    print('留言人數'+d['createdAt'])
    print('留言人數'+str(d['commentCount']))
    print('按讚人數'+str(d['likeCount']))
```

```python
# 3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」
#熱門文章的「平均留言人數」與「平均按讚人數」
r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
data = json.loads(r.text)
count=0
commentCount=0
likeCount=0
for d in data:
    count+=1
    commentCount+=d['commentCount']
    likeCount+=d['likeCount']
print('熱門文章')
print('平均留言人數:'+str(commentCount/count))
print('平均按讚人數:'+str(likeCount/count))

#非熱門文章的「平均留言人數」與「平均按讚人數」
r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=false')
data = json.loads(r.text)
count=0
commentCount=0
likeCount=0
for d in data:
    count+=1
    commentCount+=d['commentCount']
    likeCount+=d['likeCount']
print('非熱門文章')
print('平均留言人數:'+str(commentCount/count))
print('平均按讚人數:'+str(likeCount/count))

```
