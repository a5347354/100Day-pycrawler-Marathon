### 比較一下範例檔案中的「File I/O」與「xmltodict」讀出來的內容有什麼差異


File I/O讀出來是文字檔，還保有延伸式標記的符合，而xmltodict可以直結存取該element的資料

### 根據範例檔案的結果：

1. 請問高雄市有多少地區有溫度資料？
2. 請取出每一個地區所記錄的第一個時間點跟溫度
3. 請取出第一個地區所記錄的每一個時間點跟溫度

```python
# 1. 請問高雄市有多少地區有溫度資料？
import xmltodict
with open("../Data/64_72hr_CH.xml", "r") as fd:
    doc = dict(xmltodict.parse(fd.read()))
    
datasetDescription = doc['cwbopendata']['dataset']['locations']['location']
counter = 0
for data in datasetDescription:
    weathers =  data['weatherElement']
    for weather in weathers:
        if(weather['elementName'] == 'T'):
            counter = counter + 1
            
print('總共有'+str(counter)+'區')
```

```python
# 2. 請取出每一個地區所記錄的第一個時間點跟溫度
import xmltodict
with open("../Data/64_72hr_CH.xml", "r") as fd:
    doc = dict(xmltodict.parse(fd.read()))
    
datasetDescription = doc['cwbopendata']['dataset']['locations']['location']
for data in datasetDescription:
    print (data['locationName'])
    weathers =  data['weatherElement']
    for weather in weathers:
        if(weather['elementName'] == 'T'):
            print ('時間：'+weather['time'][0]['dataTime'])
            print ('溫度：'+weather['time'][0]['elementValue']['value']+weather['time'][0]['elementValue']['measures'])
```
